```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None




df = pd.read_csv('movies.csv')
```

```python
df
```

```python
# look if there are missing data

for col in df.columns:
    missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,missing))
```

```python
# delete rows that has null value

df = df.dropna(how='any',axis=0) 
```

```python
df.dtypes
```

```python
# changing column's data type

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
# creates correct year column
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df
```

```python
# look for duplicates and delete

df.drop_duplicates()
```

```python
# Check for outliers in the data

df.boxplot(column=['gross'])
```

```python
# sort by gross income

df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
# identify correlation

df.corr(method='pearson')
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix')
plt.xlabel('Movie Attrirbutes')
plt.ylabel('Movie Attrirbutes')

plt.show
```

```python
# visualize: budget vs gross revenue

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Income')
plt.xlabel('Film Budget')
plt.ylabel('Gross Income')
plt.show()
```

```python
# regression line on the scatterplot

sns.regplot(x='budget',y='gross',data=df,scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
# There is high correlation between fil, budget and gross income
```

```python
# visualize: votes vs gross income

sns.regplot(x="votes", y="gross", data=df, scatter_kws={'color':'black'}, line_kws={'color':'yellow'})
```
