```python
import pandas as pd
import numpy as np

df = pd.read_csv('estate.csv', sep=',')
df
```

```python
for x in ['price']:
    q75,q25 = np.percentile(df.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    df.loc[df[x] < min,x] = np.nan
    df.loc[df[x] > max,x] = np.nan

for x in ['bed']:
    q75,q25 = np.percentile(df.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    df.loc[df[x] < min,x] = np.nan
    df.loc[df[x] > max,x] = np.nan

for x in ['bath']:
    q75,q25 = np.percentile(df.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    df.loc[df[x] < min,x] = np.nan
    df.loc[df[x] > max,x] = np.nan

for x in ['acre_lot']:
    q75,q25 = np.percentile(df.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    df.loc[df[x] < min,x] = np.nan
    df.loc[df[x] > max,x] = np.nan

for x in ['house_size']:
    q75,q25 = np.percentile(df.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    df.loc[df[x] < min,x] = np.nan
    df.loc[df[x] > max,x] = np.nan
```

```python
# Data preprocessing for modeling
X = df.drop(['price'], axis=1).values
y = df['price'].values

# Data split
import pandas
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from warnings import simplefilter

simplefilter(action='ignore', category=FutureWarning)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=7, shuffle=True)
```

```python
# Make sure shape and data is good

print('x-train: \n',X_train)
print('y-train: \n',y_train)
print('x-test: \n',X_test)
print('y-test: \n',y_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.linear_model import BayesianRidge
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import mean_absolute_error, r2_score

# store model in variable
models = {"Linear Regression": LinearRegression(), 
          "Decision Tree": DecisionTreeRegressor(), 
          "K-Neigbors": KNeighborsRegressor(), 
          "Support Vector Machine": SVR(),
          "Bayesian Ridge Regression": BayesianRidge()
         }

for name, model in models.items():
    # Model fitting (fit train data)
    model.fit(X_train, y_train)
    
    # Test/Evaluate the model. Does it give similar/close values?
    # Predicted values per model
    y_pred = model.predict(X_test)
    
    # Compare actual y vs predicted y
    print(name, 'Actual Price:', y_test)
    print(name, 'Predicted Price:', y_pred, '\n')
```

```python
models = {"Linear Regression": LinearRegression(), 
          "Decision Tree": DecisionTreeRegressor(), 
          "K-Neigbors": KNeighborsRegressor(), 
          "Support Vector Machine": SVR(),
          "Bayesian Ridge Regression": BayesianRidge()
         }

for name, model in models.items():
    # Model fitting (fit train data)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # compute Mean Absolute Error: for every data point, how much error is there?
    print(name, "Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
    
    # compute R square/ r2 score: range from 0-1. If the model is closer to one, the better it is.
    print(name, "R-squared score:", r2_score(y_test, y_pred), '\n')
```
